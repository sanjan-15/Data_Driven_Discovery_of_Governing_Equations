```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=0.2
b=0.2
c=14

x=2
y=1
z=1
dt=0.01

t0=0
t1=50
```

```python
def derivative(x,y,z):
    return (-y-z,x+a*y,b+z*(x-c))
```

```python
X=pd.DataFrame(columns=['1','x','y','z'])
X
```

```python
while t0<=t1:
    X.loc[len(X.index)]=[1,x,y,z]

    x_,y_,z_=derivative(x,y,z)
    x+=x_*dt
    y+=y_*dt
    z+=z_*dt
    X_dot.loc[len(X_dot.index)]=[x_,y_,z_]

    t0+=dt
```
